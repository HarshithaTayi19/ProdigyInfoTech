Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\harsh\Downloads\Prodigy Internship\API_SP.POP.TOTL_DS2_en_csv_v2_5871594\API_SP.POP.TOTL_DS2_en_csv_v2_5871594.csv', skiprows=3)
```

```python
# Display the first few rows to verify the data
print(data.head())
```

```python
# Select the year 2020 and sort by population in descending order
data_2020 = data[['Country Name', '2020']].sort_values(by='2020', ascending=False).head(10)
```

```python
# Create a bar chart for the top 10 countries by population in 2020
plt.figure(figsize=(10, 6))
plt.bar(data_2020['Country Name'], data_2020['2020'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in 2020')
plt.title('Top 10 Countries by Population in 2020')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
year = '2020'
population_data = data[year]
```

```python
# Remove NaN values (if any)
population_data.dropna(inplace=True)
```

```python
# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(population_data, bins=20, edgecolor='black', color='skyblue')  # Adjust the number of bins as needed
plt.xlabel('Population')
plt.ylabel('Count')
plt.title(f'Population Distribution in {year}')
plt.tight_layout()
```
